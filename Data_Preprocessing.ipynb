```python
import csv
#raw_tweets = pd.read_csv("raw_tweets.csv")
raw_tweets= pd.read_csv("raw_tweets.csv")
columns = ["Text","Created","FollowersCount"]
raw_tweets.drop(["Unnamed: 0"],axis=1,inplace=True)
raw_tweets.columns = columns
raw_tweets
```

```python
def clean(tweet):
    tweet = re.sub(r"(https://?:\@|http?\://|https?\://|www)\S+", "", tweet)
    tweet = re.sub(r'(\s)https\w+', r'\1', tweet)
    tweet= re.sub('[0-9]+', '', tweet)
    tweet = "".join([char for char in tweet if char not in string.punctuation])
    tweet = re.sub(r"(https://?:\@|http?\://|https?\://|www)\S+", "", tweet) #Remove http links
    tweet = " ".join(tweet.split())
    tweet = ''.join(c for c in tweet if c not in emoji.UNICODE_EMOJI) #Remove Emojis
    tweet = tweet.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    return tweet
```

```python
df["Text"]= df["Text"].map(lambda x: clean(x))
```

```python
df["Created"] = pd.to_datetime(df["Created"]).dt.date
```
