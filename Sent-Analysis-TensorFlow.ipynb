```python
import pandas as pd
import numpy as np
from google.colab import files
import tensorflow as tf
import keras
from keras.preprocessing import sequence
from sklearn.model_selection import train_test_split
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
files.upload()
```

```python
data = pd.read_excel("Tweets.xlsx")
data.pop("Unnamed: 0")
```

```python
values=pd.DataFrame()
analyzer = SentimentIntensityAnalyzer()
for tweet in data["Text"]:
  sentiment = analyzer.polarity_scores(tweet)
  sents = pd.DataFrame([sentiment],columns=sentiment.keys())
  values = values.append(sents)
```

```python
values.reset_index(drop=True,inplace=True)
```

```python
df=pd.concat([data,values],axis=1)
```

```python
df
```

```python
def get_sent(compound):
  if compound > 0.05:
    return("postive")
  elif compound < -0.05:
    return ( "negative")
  else :
    return ("neutral")
```

```python
df["sentiment"] = df["compound"].map(lambda x : get_sent(x))
```

```python
df["sentiment"]
```

```python
tweet_sentiments = df[["Text","sentiment","Created"]]
```

```python
tweet_sentiments.to_csv("Tweets_sentiment.csv",index=False)
```
